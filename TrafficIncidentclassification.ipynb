```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/Crash_Incidents.csv')

# Convert Crash Date/Time to datetime format
data['Crash Date/Time'] = pd.to_datetime(data['Crash Date/Time'])

# Extract hour from Crash Date/Time
data['Hour'] = data['Crash Date/Time'].dt.hour

# Group data by Weather and Hour, and count the number of accidents
weather_hour_counts = data.groupby(['Weather', 'Hour']).size().reset_index(name='Count')

# Pivot the data to create a matrix, use keyword arguments
weather_hour_matrix = weather_hour_counts.pivot(index='Weather', columns='Hour', values='Count')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(weather_hour_matrix, cmap="YlGnBu")
plt.title('Number of Accidents by Weather and Hour of Day')
plt.show()
```
